```python
import matplotlib.pyplot as plt
```

```python
methods = ['Naive', 'im2col', 'fft']
average_time = [1.65793, 1.64817, 3.05650]  # in seconds
average_memory = [230, 310, 310]  # in bytes
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

def add_labels(ax, values):
    for i, v in enumerate(values):
        ax.text(i, v + 0.05, f'{v:.2f}', ha='center', va='bottom')

# average processing time
ax1.bar(methods, average_time, color=['cyan', 'blue', 'green'])
ax1.set_title('Average Processing Time')
ax1.set_ylabel('Time in seconds')
ax1.set_xlabel('Method')

# average memory usage
ax2.bar(methods, average_memory, color=['cyan', 'blue', 'green'])
ax2.set_title('Average Memory Usage')
ax2.set_ylabel('Memory in bytes')
ax2.set_xlabel('Method')

plt.tight_layout()
plt.show()
```
